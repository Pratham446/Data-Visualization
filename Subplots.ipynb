```python
# Ek plot ke andar multiple graph 
# Subplots allow you to show multiple plots in a single figure, side-by-side or in a grid layout. 
# This is helpful when comparing different datasets or aspects of the same data.
```

```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x] 
y2 = [i ** 2 for i in x]
y3= [i * 3 for i in x]
plt.subplot(1, 2,1)  # (rows, cols, plot_no)
plt.plot(x, y1)

plt.subplot(1, 2,2)  # (rows, cols, plot_no)
plt.plot(x, y2)

plt.subplot(1, 2,2)  # (rows, cols, plot_no)
plt.plot(x, y2)



```

```python
# 2 x 2 Subplots
y3 = [i ** 0.5 for i in x]
y4 = [10 - i for i in x]
 
# plt.figure(figsize=(8, 6))  # Optional: make it bigger
 
plt.subplot(2, 2, 1)
plt.plot(x, y1)
 
plt.subplot(2, 2, 2)
plt.plot(x, y2)

plt.subplot(2, 2, 3)
plt.plot(x, y3)

plt.subplot(2, 2, 4)
plt.plot(x, y4)
```
