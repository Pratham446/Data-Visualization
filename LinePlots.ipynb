```python
# Formmat String 
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
surya = [10, 0, 600, 700, 1120, 1300, 1600, 1700, 1800, 2000]

plt.plot(years,kohli,'ro--', label="Virat kohli")
plt.plot(years, sehwag,'gP:', label="Virender Sehwag")
plt.plot(years, surya,'b^:', label="Virender Sehwag")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.legend()
plt.show()
```

```python
# Color and Line Style Arguments
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
surya = [10, 0, 600, 700, 1120, 1300, 1600, 1700, 1800, 2000]

plt.plot(years, kohli, color='orange',linewidth=3, linestyle='--', label="Kohli")
plt.plot(years, sehwag, color='green',linewidth=2, linestyle='-.', label="Sehwag")
plt.plot(years, surya, color='blue', linestyle=':', label="surya")
plt.grid(True)
plt.legend()
plt.style.use('dark_background')
```
