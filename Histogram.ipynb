```python
# A histogram is a type of plot that shows the distribution of a dataset. Itâ€™s 
# especially useful for visualizing the frequency of numerical data within specified 
# ranges (called bins)

# Histogram is for frequency of continuous numeric data.
# or Distribution of a Data Frequency

# Bar graph is for comparison of categories.
# bins-	Number or custom edges of bins
```

```python
import matplotlib.pyplot as plt
import numpy as np
ages = [
    34, 28, 36, 45, 27, 27, 45, 37, 25, 35,
    25, 25, 32, 10, 12, 24, 19, 33, 20, 15,
    44, 27, 30, 15, 24, 31, 18, 33, 23, 27,
    23, 48, 29, 19, 38, 17, 32, 10, 16, 31,
    37, 31, 28, 26, 15, 22, 25, 40, 33, 12,
    33, 26, 23, 36, 40, 39, 21, 26, 33, 39,
    25, 28, 18, 18, 38, 43, 29, 40, 33, 23,
    33, 45, 29, 45, 3, 38, 30, 27, 30, 10,
    27, 33, 44, 24, 21, 24, 39, 33, 24, 35,
    30, 39, 22, 26, 26, 15, 32, 32, 30, 70
]
plt.hist(ages, bins=10, edgecolor='black')
plt.axvline(5, color='red', linestyle='--', linewidth=2, label='Average Age') # use to add line 
plt.axvline(np.mean(ages), color='red', linestyle='--', linewidth=2, label='Average Age') # use to find min
plt.title('Age Distribution of YouTube Viewers')
plt.xlabel('Age Group')
plt.ylabel('Number of Viewers')
plt.grid(True, linestyle='--', alpha=0.5)
```
